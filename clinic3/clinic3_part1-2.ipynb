### Question 1 Explanation – Data Leakage

Data leakage occurs when the model uses information that would not have been available at the time a prediction is made. In this dataset, the goal is to predict whether a loan will default. Therefore, any variable that describes repayment outcomes or loan balances after the loan has already been issued must be removed.

The following variables cause data leakage:

- `balance`
- `paid_total`
- `paid_principal`
- `paid_interest`
- `paid_late_fees`

These variables contain information about how much the borrower has already repaid or still owes. This information only becomes available after the loan has been issued and payments have begun. Therefore, they cannot be used as predictors when building a model that predicts default at the time of loan approval.

If these variables were kept in the dataset, the model would indirectly observe the outcome it is trying to predict. For example, borrowers who have already repaid a large amount are unlikely to default, while borrowers who have paid little may be more likely to default. As a result, the model would achieve unrealistically high performance because it would be using future information.

This phenomenon is known as **data leakage**, and it leads to overly optimistic evaluation results that would not generalize to real-world predictions.

```python
#here incude code to handle the missing values
# ================================
# Function to handle missing values
# ================================

def handle_missing_values(df):
    df_clean = df.copy()

    # Separate numeric and categorical columns
    numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
    categorical_cols = df_clean.select_dtypes(exclude=[np.number]).columns

    # Numeric: fill with median
    for col in numeric_cols:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())

    # Categorical: fill with mode (most frequent)
    for col in categorical_cols:
        # mode() can return multiple values; take the first
        mode_val = df_clean[col].mode(dropna=True)
        fill_val = mode_val.iloc[0] if len(mode_val) > 0 else "Unknown"
        df_clean[col] = df_clean[col].fillna(fill_val)

    return df_clean


# Apply and verify
loans_clean = handle_missing_values(loans)
print("Remaining missing values:", loans_clean.isnull().sum().sum())
```

### Question 2 Explanation – Missing Value Analysis

After removing the leakage variables, we analyzed the extent of missing values across the dataset.

First, we calculated the percentage of missing values for each column and identified the top 10 variables with the highest missing percentages. The visualization highlights which variables contain substantial missingness and therefore may require special handling before modeling.

Variables with more than **20% missing values** can be considered to have substantial missingness. These variables may require either imputation strategies or removal depending on their importance and predictive value.

### Missingness Patterns

Missing data can generally fall into three categories:

- **MCAR (Missing Completely at Random):** Missingness is unrelated to any other variables or the missing value itself.
- **MAR (Missing at Random):** Missingness depends on other observed variables.
- **MNAR (Missing Not at Random):** Missingness depends on the value that is missing itself.

In this dataset, several missing patterns appear to follow the **MAR mechanism**. For example, variables related to joint income (`annual_income_joint`, `verification_income_joint`) are often missing when the application type is individual rather than joint. This means the missingness is dependent on another observed variable.

Some missingness may also be **informative**. For instance, missing employment length (`emp_length`) could indicate self-employment, irregular employment, or borrowers who chose not to disclose their job history. In such cases, the missingness itself may contain useful information about borrower risk.

### Handling Missing Values

For the modeling pipeline, we implemented a simple imputation strategy:

- **Numerical variables:** missing values replaced with the median.
- **Categorical variables:** missing values replaced with the mode.

Median imputation is robust to outliers and is commonly used in financial datasets. Mode imputation allows categorical variables to retain their most common category without introducing unrealistic values.

More advanced strategies could include model-based imputation or adding indicator variables to capture whether a value was originally missing.

### Question 4 Explanation – Why Accuracy Fails

Accuracy is not an appropriate metric in this dataset because the classification problem has **highly asymmetric costs**.

In credit lending, two types of errors can occur:

**False Negative (FN):**  
A borrower who will default is predicted as safe and receives the loan.

**False Positive (FP):**  
A borrower who would repay the loan is rejected.

According to the business assumptions:

- $C_{FN} = 10{,}000$ USD (the lender loses most of the loan)  
- $C_{FP} = 500$ USD (lost profit opportunity)

This means that a **False Negative is 20 times more costly than a False Positive**.

Because of this cost imbalance, accuracy becomes misleading. A model could achieve high accuracy by predicting most borrowers as safe, but it would still fail to identify risky borrowers. Missing even a small number of defaulters can generate very large financial losses.

Therefore, evaluation metrics should focus on detecting defaults rather than simply maximizing overall accuracy.

Metrics that better reflect business risk include:

- **Recall (Sensitivity)** for the default class
- **Precision**
- **F1-score**
- **ROC-AUC**
- **Cost-based evaluation metrics**

In particular, **recall for defaulters** is important because it measures how well the model identifies borrowers who are likely to default, which directly reduces financial loss for the lender.

- the samples that are -ive and the classifier predicts (wrongly) as +ive are called False Positives (FP)
- the samples that are -ive and the classifier predicts as -ive are called True Negatives (TN)
- the samples that are +ive and the classifier predicts as -ive are called False Negatives (FN)

A classifier produces a confusion matrix which looks like this:

    *   **Recall:** 91.67%
    *   **Precision:** 80.88%
    *   **False Positives (FP):** 13

**3. Impact of Lowering the Threshold**

As we lower the predictive threshold from 0.5 to 0.2, the model requires less certainty to predict a "Default":
*   **False Positives (FP) Increase:** The number of FPs heavily increases from 1 to 13. Because the model is more aggressively looking for defaults, it mistakenly flags more actually good loans (Paid) as "Default". This causes Precision to drop from 98.1% to 80.9%.
*   **False Negatives (FN) Decrease:** The number of FNs drops from 8 to 5. The heightened sensitivity allows the model to successfully catch more of the true defaults, causing Recall to improve from 86.7% to 91.7%. 

**4. Why is 0.5 unlikely to be optimal in credit risk?**

A threshold of 0.5 implicitly assumes that False Positives and False Negatives has equal weight or penalty. In credit risk, misclassification costs are highly asymmetric:
*   A **False Negative** means the bank approves a bad loan that eventually defaults. This results in a heavy capital loss (losing the entire principal amount).
*   A **False Positive** means the bank mistakenly rejects a good loan. This only results in an opportunity cost (losing out on the potential interest margin) which is relatively small.

Because missing a true default (FN) is substantially more damaging financially than turning away a good customer (FP), banks deliberately lower their thresholds to cast a wider net. Prioritizing high Recall over mathematical Accuracy keeps the institution safe from large defaulting losses, making a 0.5 threshold less ptimal.

```python
# Your code here: Calculate and visualize if appropriate
# Question 9 Part A: Profit Optimization

# Business assumptions
PROFIT_TN = 1000    # fully paid loan approved: earn profit
LOSS_FN   = -5000   # defaulter approved: lose money
PROFIT_FP = 0       # good borrower rejected: no profit, no loss
PROFIT_TP = 0       # defaulter rejected: avoided loss, no gain

thresholds = [0.5, 0.3, 0.2]
profit_results = []

print("=" * 65)
print("PROFIT ANALYSIS AT DIFFERENT THRESHOLDS")
print("=" * 65)

for threshold in thresholds:
    y_pred_t = (y_pred_proba >= threshold).astype(int)
    cm_t = confusion_matrix(y_test, y_pred_t)
    tn_t, fp_t, fn_t, tp_t = cm_t.ravel()

    profit = tn_t * PROFIT_TN + fn_t * LOSS_FN

    profit_results.append({
        'threshold': threshold,
        'TN': tn_t, 'FP': fp_t, 'FN': fn_t, 'TP': tp_t,
        'profit': profit
    })

    print(f"\nThreshold = {threshold}")
    print(f"  TN: {tn_t:,}  FP: {fp_t:,}  FN: {fn_t:,}  TP: {tp_t:,}")
    print(f"  Profit = {tn_t:,} × $1,000 + {fn_t:,} × (−$5,000)")
    print(f"  Total Profit = ${profit:,.0f}")

# Summary table
df_profit = pd.DataFrame(profit_results)
print("\n" + "=" * 65)
print("SUMMARY")
print("=" * 65)
print(df_profit[['threshold', 'TN', 'FN', 'FP', 'TP', 'profit']].to_string(index=False))

# Bar chart
plt.figure(figsize=(6, 4))
bars = plt.bar(
    [str(t) for t in df_profit['threshold']],
    df_profit['profit'],
    color=['steelblue', 'darkorange', 'seagreen']
)
plt.title("Total Profit at Different Thresholds")
plt.xlabel("Threshold")
plt.ylabel("Total Profit (USD)")
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 5000,
             f"${bar.get_height():,.0f}",
             ha='center', fontsize=9)
plt.tight_layout()
plt.show()

#####################
#####################
```

### Question 9 Part B Answer: Fairness Challenges

#### 1. Proxy Variables for Protected Attributes

Even without explicit demographic variables like gender or race, a model may indirectly learn these protected attributes through correlated features. 

*   **Gender:** `loan_purpose` can act as a proxy. Looking at the distribution, categories like `car` or `small_business` might skew male, while `medical` or `home_improvement` might skew female. Even though `debt_consolidation` and `credit_card` dominate, the smaller categories can still introduce bias.
*   **Race / Ethnicity:** `homeownership` and geographic location variables (like `state` or `zip_code` if present) are strong proxies. Historical redlining has systematically excluded many minority groups from homeownership, linking these variables to race.
*   **Socioeconomic Status:** `annual_income` and credit limit variables directly reflect socioeconomic class, which is closely intertwined with demographic background due to wage gaps.
*   **Age:** Variables like `earliest_credit_line` or employment length correlate strongly with the age of the borrower.

The data reveals a counterintuitive pattern in default rates:

#### 3. Should We Include `annual_income`?

**Yes, but with caution.**

Income is an essential and legitimate predictor of repayment ability. The data strongly confirms this:
*   Q1 (Low Income): **5.78%** default rate
*   Q4 (High Income): **0.32%** default rate

Low-income borrowers are defaulting at nearly 18x the rate of high-income borrowers, providing a powerful predictive signal. However, due to persistent gender and racial wage gaps, treating raw income as a feature can unfairly penalize marginalized groups. A common strategy to mitigate this is to use derived metrics like the **debt-to-income ratio** rather than raw income, which focuses more specifically on financial capacity rather than absolute wealth.

#### 4. How Does Class Imbalance Interact with Fairness?

We can evaluate fairness by looking at **equal recall** across groups. Is the model equally capable of detecting defaults regardless of the borrower's background?

Our fairness check shows a major disparity based on `homeownership`:
*   RENT Recall: **92.3%** (13 defaults)
*   MORTGAGE Recall: **90.0%** (40 defaults)
*   OWN Recall: **57.1%** (7 defaults)

The model is failing to detect nearly half of the defaults for borrowers who own their homes. This is a consequence of class imbalance intersecting with group demographics:

1.  **Underrepresentation:** There are only 7 defaults in the `OWN` test group. The model lacks sufficient training examples of this specific subgroup failing to repay, meaning it hasn't learned their specific risk profile.
2.  **Imbalance Amplifies Disparities:** When we used `class_weight='balanced'`, we improved overall recall globally. However, this global improvement doesn't distribute evenly. The dominant groups (like MORTGAGE and RENT) received the benefit, while the underrepresented OWN group was left behind, making the model potentially *less fair* across subgroups even as it became more accurate overall.

#### 5. Mitigation Strategies

*   **Strategy 1 (In-Processing): Fairness-Aware Instance Reweighting:** Instead of just reweighting based on the global target class (Default vs. Paid), we can assign weights based on the *intersection* of the class and the demographic group. A rare case (e.g., an `OWN` borrower who defaults) gets a significantly higher training weight, forcing the model to pay attention to that specific profile.
*   **Strategy 2 (Post-Processing): Group-Specific Thresholds:** Because the model struggles to detect defaults in the `OWN` group (leading to low recall), we could apply a lower probability threshold for predicting "Default" specifically for that group, while maintaining higher thresholds for groups where the model is more confident. This helps equalize the recall across all demographics.
